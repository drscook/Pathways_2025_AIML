```python
# prompt: do a 80-20 train-test split stratified by survived, sex, and class

from sklearn.model_selection import train_test_split

# Stratified split by 'survived', 'sex', and 'class'
X = titanic_df.drop('survived', axis=1)
y = titanic_df['survived']
stratify_cols = titanic_df[['survived', 'sex', 'class']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = stratify_cols, random_state=42)

print("Train set size:", len(X_train))
print("Test set size:", len(X_test))
```

```python
# prompt: confirm that stratification was successfull

import pandas as pd
import numpy as np
# Verify stratification
def verify_stratification(df, train_indices, test_indices, stratify_cols):
    train_df = df.iloc[train_indices]
    test_df = df.iloc[test_indices]

    for col in stratify_cols:
        train_dist = train_df[col].value_counts(normalize=True).sort_index()
        test_dist = test_df[col].value_counts(normalize=True).sort_index()

        print(f"Stratification check for column: {col}")
        print("Train set distribution:")
        print(train_dist)
        print("Test set distribution:")
        print(test_dist)

        # Check if distributions are reasonably close
        if not np.allclose(train_dist.values, test_dist.values, atol=0.05):
            print(f"WARNING: Stratification might not be perfect for {col}. Distributions are not very close.")
        else:
            print(f"Stratification looks good for {col}.")
        print("-" * 40)

# Convert stratify_cols to a list of column names if it's not already
if isinstance(stratify_cols, pd.DataFrame):
    stratify_cols = stratify_cols.columns.tolist()


# Get indices from train_test_split
train_indices = X_train.index
test_indices = X_test.index

# Verify stratification
verify_stratification(titanic_df, train_indices, test_indices, stratify_cols)
```

```python
# prompt: create a random forest classifier where survived is the target variable

import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Handle missing values by filling with the mean for numerical columns
for col in X_train.select_dtypes(include=np.number).columns:
    X_train[col] = X_train[col].fillna(X_train[col].mean())
    X_test[col] = X_test[col].fillna(X_train[col].mean())  # Use training mean for test set

# Initialize and train the Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

```python
# prompt: show the confusion matrix for the test set

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Survived', 'Survived'],
            yticklabels=['Not Survived', 'Survived'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# prompt: show the ROC curve

import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Get predicted probabilities for the positive class
y_prob = model.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

```python
# prompt: try the same supervised task using other classifiers and tell me which has the best accuracy

import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline


# Define a dictionary of classifiers
classifiers = {
    "Logistic Regression": LogisticRegression(random_state=42, solver='liblinear'),
    "Support Vector Machine": SVC(random_state=42, probability=True),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Neural Network": MLPClassifier(random_state=42, max_iter=300)
}

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train and evaluate each classifier
results = {}
for name, clf in classifiers.items():
    # Handle classifiers that need a scaled input
    if name in ["Logistic Regression", "Support Vector Machine", "K-Nearest Neighbors","Neural Network"]:
      pipeline = Pipeline([('scaler', StandardScaler()), ('clf', clf)])
      pipeline.fit(X_train, y_train)
      y_pred = pipeline.predict(X_test)
    else:
      clf.fit(X_train, y_train)
      y_pred = clf.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy
    print(f"{name} Accuracy: {accuracy:.4f}")

# Find the best classifier
best_classifier = max(results, key=results.get)
best_accuracy = results[best_classifier]
print("-" * 40)
print(f"Best Classifier: {best_classifier} with Accuracy: {best_accuracy:.4f}")
print("-" * 40)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.bar(results.keys(), results.values(), color='skyblue')
plt.xlabel('Classifier')
plt.ylabel('Accuracy')
plt.title('Classifier Accuracy Comparison')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```
